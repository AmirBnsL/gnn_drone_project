```python


path = "../datasets/formation_triangle_comm_10m_triangle.npz"
data = np.load(path, allow_pickle=True)

# dig into states
print("=== STATES ===")
print("list:", type(data['states'][0]))           # list
print("drone state:", type(data['states'][0][0]))        # what is ONE drone's state?
print("drone 0 at step 0:", data['states'][0][0])             # print drone 0 at step 0
print("num drones at step 0:", len(data['states'][0]))           # how many drones at step 0?
print("features per drone:", len(data['states'][0][0]))        # how many features per drone?

# dig into targets
print("\n=== TARGETS ===")
print("drone target:", type(data['targets'][0][0]))       # what is ONE drone's target?
print("drone 0 target at step 0:", data['targets'][0][0])            # print drone 0 target at step 0
print("target features:", len(data['targets'][0][0]))       # should be 4

# dig into edges
print("\n=== EDGES ===")
print("edge:", type(data['edges'][0][0]))         # what is ONE edge?
print("first edge:", data['edges'][0][0])              # print first edge
print("num edges at step 0:", len(data['edges'][0]))            # number of edges at step 0

# check a few steps to see if num drones changes
print("\n=== NUM DRONES PER STEP ===")
for i in [0, 1, 2, 100, 500]:
    print(f"step {i}: {len(data['states'][i])} drones, {len(data['edges'][i])} edges")
```

```python


path = "../datasets/formation_triangle_comm_10m_triangle.npz"
data = np.load(path, allow_pickle=True)

# How many episodes, how many drones per episode?
print(f"Total timesteps     : {len(data['states'])}")
print(f"Drones at step 0    : {len(data['states'][0])}")
print(f"Drones at step 500  : {len(data['states'][500])}")
print(f"Drones at step 4000 : {len(data['states'][4000])}")

# Check targets — are they varied or all zeros?
import numpy as np
all_targets = np.array([data['targets'][i][0] for i in range(100)])
print(f"\nTarget stats over first 100 steps, drone 0:")
print(f"  mean : {all_targets.mean(axis=0).round(3)}")
print(f"  std  : {all_targets.std(axis=0).round(3)}")
print(f"  min  : {all_targets.min(axis=0).round(3)}")
print(f"  max  : {all_targets.max(axis=0).round(3)}")

# Check formations
unique_formations = set(data['formation_names'])
print(f"\nFormations in dataset: {unique_formations}")
print(f"Formation IDs sample : {data['formation_ids'][:5]}")
```
